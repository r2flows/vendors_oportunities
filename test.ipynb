```python
def clasificar_productos(df_analisis_precios):
    """
    Clasifica los productos en tres categorías basadas en su diferencia porcentual de precio
    y genera resúmenes estadísticos.
    
    Args:
        df_analisis_precios (pd.DataFrame): DataFrame con el análisis de precios original
        
    Returns:
        tuple: (df_productos_clasificados, resumen_clasificacion, resumen_vendor_clasificacion)
    """
    # 1. Primero creamos los subsets por categoría
    productos_5_porciento = df_analisis_precios[
        df_analisis_precios['categoria_diferencia'] == 'hasta_5_porciento'
    ].copy()

    productos_15_porciento = df_analisis_precios[
        df_analisis_precios['categoria_diferencia'] == 'hasta_15_porciento'
    ].copy()

    productos_mayor_15 = df_analisis_precios[
        df_analisis_precios['categoria_diferencia'] == 'mayor_15_porciento'
    ].copy()

    # 2. Añadimos la nueva clasificación a cada subset
    productos_5_porciento['clasificacion_producto'] = 'Producto Ganador'
    productos_15_porciento['clasificacion_producto'] = 'Producto Medianamente Ganador'
    productos_mayor_15['clasificacion_producto'] = 'Producto No Ganador'

    # 3. Concatenamos los DataFrames
    df_productos_clasificados = pd.concat([
        productos_5_porciento,
        productos_15_porciento,
        productos_mayor_15
    ], ignore_index=True)

    # 4. Reordenamos las columnas para mejor visualización
    columnas_orden = [
        'super_catalog_id',
        'point_of_sale_id',
        'vendor_id',
        'clasificacion_producto',
        'categoria_diferencia',
        'diferencia_porcentual',
        'precio_vendedor',
        'precio_minimo_orders',
        'unidades_pedidas',
        'geo_zone'
    ]

    df_productos_clasificados = df_productos_clasificados[columnas_orden]

    # 5. Agregamos algunos resúmenes útiles
    display("\nResumen de productos por clasificación:")
    resumen_clasificacion = df_productos_clasificados.groupby('clasificacion_producto').agg({
        'super_catalog_id': 'count',
        'diferencia_porcentual': ['mean', 'min', 'max'],
        'unidades_pedidas': 'sum'
    }).round(2)

    display("\nResumen por vendor y clasificación:")
    resumen_vendor_clasificacion = df_productos_clasificados.groupby(
        ['vendor_id', 'clasificacion_producto']
    ).agg({
        'super_catalog_id': 'count',
        'diferencia_porcentual': 'mean',
        'unidades_pedidas': 'sum'
    }).round(2)

    # 6. Exportamos el DataFrame clasificado a CSV
    df_productos_clasificados.to_csv('productos_clasificados.csv', index=False, encoding='utf-8')
    display("\nArchivo 'productos_clasificados.csv' ha sido creado exitosamente.")
    
    return df_productos_clasificados, resumen_clasificacion, resumen_vendor_clasificacion
```

```python
from sshtunnel import SSHTunnelForwarder
import pandas as pd
import paramiko
from io import StringIO
import pymysql
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
ssh_host = os.getenv('SSH_HOST')
ssh_port = 22
ssh_user = os.getenv('SSH_USER')
ssh_pkey= '/home/dell/.ssh/id_rsa'
ssh_passphrase = 'hacker' # Reemplaza con tu passphrase si es necesario

# Parámetros de la base de datos
db_host = os.getenv("DB_HOST")
db_port = 3306
db_user = os.getenv("DB_USER")
db_password = os.getenv("DB_PASSWORD")
db_name = os.getenv("DB_NAME")

# Cargar la clave privada desde el objeto en memoria
my_pkey = paramiko.RSAKey.from_private_key_file(ssh_pkey, password=ssh_passphrase)

# Configurar la conexión SSH
client = paramiko.SSHClient()
client.set_missing_host_key_policy(paramiko.AutoAddPolicy())
client.connect(hostname=ssh_host, port=ssh_port, username=ssh_user, pkey=my_pkey)
```

```python
def vendor_prices():
    with SSHTunnelForwarder(
        (ssh_host, ssh_port),
        ssh_username=ssh_user,
        ssh_pkey=my_pkey,  # Usar la clave privada de Paramiko
        remote_bind_address=(db_host, db_port)
    ) as tunnel:
        # Conectar a la base de datos usando pymysql
        connection = pymysql.connect(
            host='127.0.0.1',
            port=tunnel.local_bind_port,
            user=db_user,
            password=db_password,
            db=db_name
        )
        try:
            with connection.cursor() as cursor:
                # Especificar la consulta SQL
                query = """
SELECT vvc.vendor_id, 
vvc.super_catalog_id, 
vsz.name, 
vvc.base_price, 
vp.percentage
FROM v_vendor_catalog vvc 
join clients c on vvc.vendor_id = c.id and c.is_demo = 0 
JOIN v_prices vp on vvc.id = vp.product_id and vp.deleted_at is null 
and vp.min_price is null and vp.max_price is null
join v_price_list vpl on vpl.id = vp.price_list_id and vpl.deleted_at is null and vpl.status = 1 
JOIN v_shipping_zones vsz on vpl.delivery_zone_id = vsz.id 
WHERE vvc.deleted_at is null

                """

                # Ejecutar la consulta
                cursor.execute(query)
                
                # Obtener los resultados
                results = cursor.fetchall()

                # Convertir los resultados a un DataFrame de pandas
                v_prices = pd.DataFrame(results, columns=[
                'vendor_id', 
                'super_catalog_id',
                'geo_zone',
                'base_price',
                'percentage'
                ])
        except Exception as e:
            print(f"Error: {e}")

        finally:
            connection.close()

    return v_prices
```

```python
import pandas as pd
import numpy as np

def swap_price_bands(merged):
    """
    Intercambia los valores de min_price y max_price en los casos donde min_price > max_price
    
    Args:
        df (pd.DataFrame): DataFrame con las columnas min_price y max_price
        
    Returns:
        pd.DataFrame: DataFrame con los precios corregidos
    """
    # Crear una copia del dataframe para no modificar el original
    df_fixed =merged.copy()
    
    # Identificar casos donde min > max
    mask_invalid = merged['min_price'] > merged['max_price']
    
    # Guardar temporalmente los valores originales
    original_min = df_fixed.loc[mask_invalid, 'min_price'].copy()
    
    # Hacer el intercambio para los casos inválidos
    df_fixed.loc[mask_invalid, 'min_price'] = df_fixed.loc[mask_invalid, 'max_price']
    df_fixed.loc[mask_invalid, 'max_price'] = original_min
    df_fixed['min_price'] = df_fixed['min_price']*0.99
    return df_fixed

def analyze_changes(df_original, df_fixed):
    """
    Analiza los cambios realizados en las bandas de precios
    
    Args:
        df_original (pd.DataFrame): DataFrame original
        df_fixed (pd.DataFrame): DataFrame con precios corregidos
        
    Returns:
        pd.DataFrame: DataFrame con el análisis de cambios
    """
    # Identificar casos que fueron corregidos
    mask_corrected = df_original['min_price'] > df_original['max_price']
    
    # Crear DataFrame con la comparación
    comparison = pd.DataFrame({
        'super_catalog_id': df_original.loc[mask_corrected, 'super_catalog_id'],
        'vendor_id': df_original.loc[mask_corrected, 'vendor_id'],
        'original_min': df_original.loc[mask_corrected, 'min_price'],
        'original_max': df_original.loc[mask_corrected, 'max_price'],
        'new_min': df_fixed.loc[mask_corrected, 'min_price'],
        'new_max': df_fixed.loc[mask_corrected, 'max_price']
    })
    
    # Calcular diferencias
    comparison['original_diff'] = comparison['original_min'] - comparison['original_max']
    comparison['new_diff'] = comparison['new_max'] - comparison['new_min']
    
    return comparison

if __name__ == "__main__":
    # 1. Leer el archivo
    #merged = pd.read_csv('banda_precios_actualizada.csv')
    
    # 2. Aplicar la corrección
    df_fixed = swap_price_bands(merged)
    display(df_fixed)
    # 3. Analizar los cambios
    comparison = analyze_changes(merged, df_fixed)
    
    # 4. Mostrar estadísticas
    total_cases = len(comparison)
    print(f"\nTotal de casos corregidos: {total_cases}")
    
    # Mostrar algunos ejemplos
    print("\nEjemplos de correcciones (primeros 5 casos):")
    print(comparison.head().round(2))
    
    # Verificar que no quedan casos inválidos
    invalid_after = (df_fixed['min_price'] > df_fixed['max_price']).sum()
    print(f"\nCasos inválidos después de la corrección: {invalid_after}")
    
    # 5. Guardar resultados
    df_fixed.to_csv('banda_precios_actualizado.csv', index=False)
    comparison.to_csv('reporte_correcciones.csv', index=False)
```
