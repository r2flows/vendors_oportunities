```python
import pandas as pd
import numpy as np

def analyze_price_differences(vendors_catalog_path, min_prices_path):
    # Leer los archivos CSV
    min_prices_df = pd.read_csv(min_prices_path)
    vendors_catalog_df = pd.read_csv(vendors_catalog_path)
    
    # Crear precio de venta en el catálogo de vendors con la fórmula corregida
    vendors_catalog_df['precio_vendedor'] = vendors_catalog_df['base_price'] + (vendors_catalog_df['base_price'] * vendors_catalog_df['percentage']/100)
    
    # Obtener precios mínimos únicos por producto
    min_prices_reference = min_prices_df[['super_catalog_id', 'precio_minimo']].drop_duplicates()
    
    # Unir con el catálogo de vendors
    df_comparacion = pd.merge(
        vendors_catalog_df,
        min_prices_reference,
        on='super_catalog_id',
        how='inner'
    )
    
    # Calcular diferencia porcentual
    df_comparacion['diferencia_porcentual'] = (
        (df_comparacion['precio_vendedor'] - df_comparacion['precio_minimo']) / 
        df_comparacion['precio_minimo'] * 100
    )
    
    # Agregar columna de diferencia absoluta
    df_comparacion['diferencia_absoluta'] = df_comparacion['diferencia_porcentual'].abs()
    
    # Categorizar las diferencias usando el valor absoluto
    def categorizar_diferencia(diferencia_abs):
        if diferencia_abs <= 5:
            return 'hasta_5_porciento'
        elif diferencia_abs <= 15:
            return 'hasta_15_porciento'
        else:
            return 'mayor_15_porciento'
    
    df_comparacion['categoria_diferencia'] = df_comparacion['diferencia_absoluta'].apply(categorizar_diferencia)
    
    # Crear resúmenes
    resumen_por_categoria = df_comparacion.groupby('categoria_diferencia').agg({
        'super_catalog_id': 'count',
        'vendor_id': lambda x: len(x.unique())
    }).reset_index()
    
    resumen_por_categoria.columns = [
        'categoria_diferencia',
        'cantidad_productos',
        'cantidad_vendors'
    ]
    
    resumen_por_vendor = df_comparacion.groupby(['vendor_id', 'categoria_diferencia']).agg({
        'super_catalog_id': 'count',
        'diferencia_porcentual': ['mean', 'min', 'max'],
        'diferencia_absoluta': 'mean'
    }).reset_index()
    
    resumen_por_vendor.columns = [
        'vendor_id',
        'categoria_diferencia',
        'cantidad_productos',
        'diferencia_promedio',
        'diferencia_minima',
        'diferencia_maxima',
        'diferencia_absoluta_promedio'
    ]
    
    # Identificar productos más representativos por categoría
    productos_representativos = df_comparacion.sort_values('diferencia_absoluta').groupby('categoria_diferencia').head(5)
    
    return {
        'detalle_completo': df_comparacion,
        'resumen_por_categoria': resumen_por_categoria,
        'resumen_por_vendor': resumen_por_vendor,
        'productos_representativos': productos_representativos[['super_catalog_id', 'vendor_id', 
                                                             'precio_vendedor', 'precio_minimo', 
                                                             'diferencia_porcentual', 'diferencia_absoluta',
                                                             'categoria_diferencia']]
    }

# Ejemplo de uso:
resultados = analyze_price_differences('vendors_catalog.csv', 'top_5_productos_geozona.csv')

# Mostrar resultados
display("\nResumen por categoría:")
display(resultados['resumen_por_categoria'])

display("\nEjemplos de productos representativos por categoría:")
for categoria in ['hasta_5_porciento', 'hasta_15_porciento', 'mayor_15_porciento']:
    display(f"\n{categoria}:")
    subset = resultados['productos_representativos'][
        resultados['productos_representativos']['categoria_diferencia'] == categoria
    ]
    display(subset[['super_catalog_id', 'vendor_id', 'precio_vendedor', 
                 'precio_minimo', 'diferencia_porcentual', 'diferencia_absoluta']])
```

```python
def clasificar_productos(df_analisis_precios):
    """
    Clasifica los productos en tres categorías basadas en su diferencia porcentual de precio
    y genera resúmenes estadísticos.
    
    Args:
        df_analisis_precios (pd.DataFrame): DataFrame con el análisis de precios original
        
    Returns:
        tuple: (df_productos_clasificados, resumen_clasificacion, resumen_vendor_clasificacion)
    """
    # 1. Primero creamos los subsets por categoría
    productos_5_porciento = df_analisis_precios[
        df_analisis_precios['categoria_diferencia'] == 'hasta_5_porciento'
    ].copy()

    productos_15_porciento = df_analisis_precios[
        df_analisis_precios['categoria_diferencia'] == 'hasta_15_porciento'
    ].copy()

    productos_mayor_15 = df_analisis_precios[
        df_analisis_precios['categoria_diferencia'] == 'mayor_15_porciento'
    ].copy()

    # 2. Añadimos la nueva clasificación a cada subset
    productos_5_porciento['clasificacion_producto'] = 'Producto Ganador'
    productos_15_porciento['clasificacion_producto'] = 'Producto Medianamente Ganador'
    productos_mayor_15['clasificacion_producto'] = 'Producto No Ganador'

    # 3. Concatenamos los DataFrames
    df_productos_clasificados = pd.concat([
        productos_5_porciento,
        productos_15_porciento,
        productos_mayor_15
    ], ignore_index=True)

    # 4. Reordenamos las columnas para mejor visualización
    columnas_orden = [
        'super_catalog_id',
        'point_of_sale_id',
        'vendor_id',
        'clasificacion_producto',
        'categoria_diferencia',
        'diferencia_porcentual',
        'precio_vendedor',
        'precio_minimo_orders',
        'unidades_pedidas',
        'geo_zone'
    ]

    df_productos_clasificados = df_productos_clasificados[columnas_orden]

    # 5. Agregamos algunos resúmenes útiles
    display("\nResumen de productos por clasificación:")
    resumen_clasificacion = df_productos_clasificados.groupby('clasificacion_producto').agg({
        'super_catalog_id': 'count',
        'diferencia_porcentual': ['mean', 'min', 'max'],
        'unidades_pedidas': 'sum'
    }).round(2)

    display("\nResumen por vendor y clasificación:")
    resumen_vendor_clasificacion = df_productos_clasificados.groupby(
        ['vendor_id', 'clasificacion_producto']
    ).agg({
        'super_catalog_id': 'count',
        'diferencia_porcentual': 'mean',
        'unidades_pedidas': 'sum'
    }).round(2)

    # 6. Exportamos el DataFrame clasificado a CSV
    df_productos_clasificados.to_csv('productos_clasificados.csv', index=False, encoding='utf-8')
    display("\nArchivo 'productos_clasificados.csv' ha sido creado exitosamente.")
    
    return df_productos_clasificados, resumen_clasificacion, resumen_vendor_clasificacion
```
